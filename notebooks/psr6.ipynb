```python
to_encrypt = 1029830129
encryption_key = dict(zip(list(range(10)), list(range(9, -1, -1))))
decryption_key = dict(zip(encryption_key.values(), encryption_key.keys()))
print(encryption_key)
print(decryption_key)

encrypted = ''
for i in str(to_encrypt):
  encrypted += str(encryption_key[int(i)])
print(int(encrypted))

decrypted = ''
for i in encrypted:
  decrypted += str(decryption_key[int(i)])
print(int(decrypted))

print(to_encrypt == int(decrypted))
```
