## Calculate Body Mass Index

```
bmi = weight / (height ** 2)
```

````python
def calculate_bmi(weight, height):
  # remember, height in cm!
  height_m = height / 100
  bmi = weight / (height_m ** 2)
  return bmi

calculate_bmi(81, 182)
````

Calculate BMI. If `output` is None (Default), return raw score. If `output` is "categories", return weight categories:

- Underweight: 18.5 >= bmi
- Healthy: 18.5 < bmi < 25
- Overweight: 25 <= bmi < 30
- Obesity: bmi >= 30



````python
def calculate_bmi(weight, height, output=None):
  # remember, height in cm!
  height_m = height / 100
  bmi = weight / (height_m ** 2)
  if output is None:
    return bmi

  elif output == "categories":
    if bmi <= 18.5:
      return 'underweight'
    elif bmi < 25:
      return 'healthy'
    elif bmi < 30:
      return 'overweight'
    else:
      return 'obese'

calculate_bmi(81, 183, 'categories')
````

````python
calculate_bmi(height=160, weight=40, output=None)
````
