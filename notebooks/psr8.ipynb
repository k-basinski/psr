Write a function that takes a list of integer numbers and for each element checks if the element is odd or even. Return a list of boolean values that are True if element is odd and False if even.

````python
def is_odd(x):
  if x % 2 == 0:
    return False
  else:
    return True


def is_list_odd(no_list):
  result = []
  for elem in no_list:
    result.append(is_odd(elem))
  return result

is_list_odd(l)
````

````python
def check_odd_numbers(lst):
   res = [is_odd(num) for num in lst]
   return res

check_odd_numbers(l)
````

````python
is_list_odd(l) == check_odd_numbers(l)
````

Assignment: Write a function that calculates the n-th element of the fibonacci sequence:

````python
def fib(n):
  if n <= 1:
    return n
  else:
    return fib(n - 1) + fib(n - 2)

[fib(n) for n in range(1,10)]
fib(10)
````

````python
def fib(n):
  seq = [1, 2]
  for i in range(3, n):
    seq.append(seq[-1] + seq[-2])
  return seq[-1]

fib(10)
````
