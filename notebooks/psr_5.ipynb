Write a `while` loop that prints all numbers NOT divisible by 2 or 3, or 4 in the range of 1 to 100.

```python
b = 1
while b <= 100:
  if b % 2 != 0 and b % 3 != 0 and b % 4 != 0:
    print(b)
  b = b + 1
```

```python
i = 1
while i < 101:
  if (i % 2 != 0) and (i % 3 != 0) and (i % 4 != 0):
    print(i)
  i += 1
```

## `for` loops

They're made to iterate over stuff. They work with _iterable_ objects.

```python
smth = 'something'
for s in smth:
  print(s)
```

```python
for letter in smth:
  print(letter)
```

```python
veggies = ['potato', 'cucumber', 'beetroot', 'onion', 'carrot']

for veggie in veggies:
  print(f'In our store, we have some {veggie}s.')
```

```python
for v in veggies:
  upper_v = v.upper()
  print(f'In our store, we have some {upper_v}s.')
```

```python
stock = [3, 1, 10, 5, 5]

print(f'There are {stock[0]} kg of {veggies[0]}.')
```

```python
i = 0
while i < len(veggies):
  print(f'There are {stock[i]} kg of {veggies[i]}.')
  i += 1
```

```python
for i, veg in enumerate(veggies):
  print(f'{i}. There are {stock[i]} kg of {veg}.')
```

```python
for veg, st in zip(veggies, stock):
  print(f'There are {st} kg of {veg}.')
```
