```python
fruit = ['banana', 'apple', 'orange']
```

```python
words = ['tree', 'car', 'banana', 'apple', 'cucumber', 'orange', 'guitar', 'piano', 'computer', 'camera']
```

```python
'trunk' in words
```

```python
# x in y
# how many fruit in words?
i = 0
for w in words:
  if w in fruit:
    i += 1

i
```

```python
def how_many_elements(x, targets):
  i = 0
  for w in x:
    if w in targets:
      i += 1

  return i


how_many_elements(words, fruit)
```

```python
how_many_elements(['aplle', 'apple', 'aaple'], fruit)
```

```python
how_many_elements([1, 2, 3, 4,4, 4, 4, 4], [4])
```
